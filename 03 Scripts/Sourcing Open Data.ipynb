```python
import pandas as pd

# Load the dataset
file_path = r'C:\Users\sudee\OneDrive\Documents\Newyork Citi Bike Project\02 Data\Original Data\citibike.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())

# Basic data cleaning
# Remove duplicates
df = df.drop_duplicates()

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Optionally, drop rows with missing values or handle them appropriately
df = df.dropna()  # Dropping rows with missing values (if applicable)

# Convert any necessary columns to the correct data type (e.g., dates)
# Assuming 'starttime' and 'stoptime' are datetime fields
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# Display cleaned data information
print(df.info())

# Save the cleaned data (optional)
df.to_csv(r'C:\Users\sudee\OneDrive\Documents\Newyork Citi Bike Project\02 Data\prepared Data\cleaned_citibike.csv', index=False)
```

```python
# Descriptive statistics for continuous variables
descriptive_stats = df.describe()
print("Descriptive Statistics for Continuous Variables:")
print(descriptive_stats)

# Frequency counts for categorical variables
categorical_columns = ['weekday', 'subscriber', 'gender']  # Assuming these are categorical columns in your data
for col in categorical_columns:
    print(f"\nValue counts for {col}:\n", df[col].value_counts())
```
